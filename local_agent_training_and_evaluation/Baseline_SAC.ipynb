```python
import gym
import numpy as np
```

```python
env = gym.make(
    "reference_environment:rangl-nztc-v0"
)
env.param.noise_observability = True
```

```python
env.observation_space
```

```python
from stable_baselines3.sac.policies import MlpPolicy
from stable_baselines3 import SAC
```

```python
np.array(env.reset())
```

```python
from gym import Wrapper, spaces
```

```python
class numpyWrapper(Wrapper):
    def __init__(self, env):
        super(numpyWrapper, self).__init__(env)
        
    def reset(self):
        state = self.env.reset()
        self.numpy_state = np.array(state)
        return self.numpy_state
        
    def step(self, action, render=False):
        new_state, reward, done, info = self.env.step(action)
        self.numpy_state = np.array(new_state)
        return self.numpy_state, reward, done, info
```

```python
env.reset()
env_numpy = numpyWrapper(env)
```

```python
env_numpy.reset()
env_numpy.step([0,0,0])
```

```python
model = SAC(MlpPolicy, env_numpy, verbose=1)
model.learn(total_timesteps=10000, log_interval=10)
```

```python
PROVE = 100
returns = np.zeros(PROVE)


for i in range(PROVE):
    
    obs = env_numpy.reset()
    done = False
    ret = 0

    while(not done):
        action, _states = model.predict(obs)        
        new_obs, reward, done, info = env_numpy.step(action)
        ret += reward
        obs = new_obs
    returns[i] = ret  

print(str(np.mean(returns))+"+-"+str(np.std(returns)))
```
